```python
import sys
sys.path.append('..')

from src.mortgage_calculator import MortgageCalculator
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set style
plt.style.use('seaborn-v0_8-whitegrid') # to view seaborn styles: [x for x in plt.style.available if 'seaborn' in x]
sns.set_palette('husl')
```

```python
# Initialize calculator
calc = MortgageCalculator()

# Analyze scenarios for a £500,000 property
scenarios_df = calc.analyze_scenarios(property_value=500000, years=25)  # Assuming 25 years as the mortgage duration

# Display results
print("Mortgage scenarios for a £500,000 property:")
display(scenarios_df)
```

```python
# Visualize deposit requirements and monthly payments
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot deposit requirements
sns.barplot(data=scenarios_df, x='LTV', y='Deposit', ax=ax1)
ax1.set_title('Required Deposit by LTV')
ax1.set_ylabel('Deposit (£)')

# Plot monthly payments range
scenarios_df.plot(x='LTV', y=['Best Monthly', 'Lower Mid Monthly', 
                             'Higher Mid Monthly', 'Worst Monthly'], 
                 kind='bar', ax=ax2, width=0.8)
ax2.set_title('Monthly Payment Ranges by LTV')
ax2.set_ylabel('Monthly Payment (£)')
ax2.legend(title='Payment Scenarios')

plt.tight_layout()
plt.show()
```

```python
# Initialize calculator
calc = MortgageCalculator()

# Define property values to analyze (in £)
property_values = [400_000, 500_000, 750_000, 1_000_000]

# Analyze scenarios for all property values
all_scenarios = {}
for value in property_values:
    all_scenarios[value] = calc.analyze_scenarios(property_value=value, years=25)

# Create a combined DataFrame with property value as an additional column
combined_df = pd.concat([df.assign(Property_Value=value) 
                        for value, df in all_scenarios.items()], 
                        ignore_index=True)

# Display summary
print("Mortgage scenarios summary:")
display(combined_df)

# Visualizations
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
axes = axes.ravel()

# 1. Deposit Requirements
sns.barplot(data=combined_df, 
           x='LTV', 
           y='Deposit',
           hue='Property_Value',
           ax=axes[0])
axes[0].set_title('Required Deposit by LTV and Property Value')
axes[0].set_ylabel('Deposit (£)')
axes[0].tick_params(axis='x', rotation=45)

# 2. Best Monthly Payments
sns.barplot(data=combined_df,
           x='LTV',
           y='Best Monthly',
           hue='Property_Value',
           ax=axes[1])
axes[1].set_title('Best Case Monthly Payments')
axes[1].set_ylabel('Monthly Payment (£)')
axes[1].tick_params(axis='x', rotation=45)

# 3. Payment Ranges for each Property Value
for idx, value in enumerate(property_values):
    data = all_scenarios[value]
    spread = data['Worst Monthly'] - data['Best Monthly']
    axes[2].scatter(data['LTV'], data['Best Monthly'], label=f'£{value:,}')
    axes[2].fill_between(data['LTV'], 
                        data['Best Monthly'], 
                        data['Worst Monthly'], 
                        alpha=0.2)
axes[2].set_title('Monthly Payment Ranges')
axes[2].set_xlabel('LTV (%)')
axes[2].set_ylabel('Monthly Payment (£)')
axes[2].legend(title='Property Value')

# 4. Rate Spreads
for value in property_values:
    data = all_scenarios[value]
    axes[3].plot(data['LTV'], data['Best Rate'], 
                marker='o', label=f'£{value:,} Best')
    axes[3].plot(data['LTV'], data['Worst Rate'], 
                marker='o', linestyle='--', label=f'£{value:,} Worst')
axes[3].set_title('Interest Rate Ranges by LTV')
axes[3].set_xlabel('LTV (%)')
axes[3].set_ylabel('Interest Rate (%)')
axes[3].legend(title='Property Value & Scenario')

plt.tight_layout()
plt.show()

# Additional analysis: Cost comparison table
comparison_df = combined_df.copy()
comparison_df['Total_Best_Cost'] = comparison_df['Best Monthly'] * 12 * 25
comparison_df['Total_Worst_Cost'] = comparison_df['Worst Monthly'] * 12 * 25
comparison_df['Cost_Difference'] = comparison_df['Total_Worst_Cost'] - comparison_df['Total_Best_Cost']

summary_table = comparison_df.groupby('Property_Value').agg({
    'Total_Best_Cost': 'mean',
    'Total_Worst_Cost': 'mean',
    'Cost_Difference': 'mean',
    'Best Monthly': 'mean',
    'Worst Monthly': 'mean'
}).round(2)

print("\nAverage Costs Summary:")
display(summary_table)
```
